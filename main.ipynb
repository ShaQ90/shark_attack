```python
import pandas as pd
import numpy as np
import re
import my_functions as mf
from datetime import datetime
import plotly.express as px


#define were the data comes from
df = pd.read_excel("https://www.sharkattackfile.net/spreadsheets/GSAF5.xls")
```

```python
#cleaning the data as we see fit

df = mf.cleaning_tool(df)
#reseting the index
df = df.reset_index()
```

```python
#adding new column for time classification

df["Time_Classification"] = df["Time"].apply(mf.classify_time)
```

```python
#adding new column for if atack was fatal
pattern = r'\b[fF][aA][tT][aA][lL]\b'
df['Fatal'] =  df['Injury'].str.contains(pattern, regex = True)
```

```python
#Convert date as date type
df["Date"] = df["Date"].apply(mf.convert_to_dates)
```
